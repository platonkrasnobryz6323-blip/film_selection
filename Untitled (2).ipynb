```python
# --- 1. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫ ---
import numpy as np
import pandas as pd # –î–ª—è —Ä–∞–±–æ—Ç—ã —Å –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞–º–∏
from sklearn.neighbors import NearestNeighbors # –î–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –ª–æ–≥–∏–∫–∏ KNN
from sklearn.preprocessing import StandardScaler # –î–ª—è –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö

print("–ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã.")
```

```python
# --- 2. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö ---
file_path = '/IMDb_All_Genres_etf_clean1.csv'

try:
    df = pd.read_csv(file_path)
    print("–î–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω—ã! –†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞:", df.shape)
    print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞—Ç–∞—Å–µ—Ç–∞:")
    print(df.head())
    print("\n–î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏:", list(df.columns))
except FileNotFoundError:
    print("‚ö†Ô∏è –û—à–∏–±–∫–∞: –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω. –ü—Ä–æ–≤–µ—Ä—å –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É!")
```

```python
# --- 3. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö (One-Hot Encoding) ---
# –í—ã–±–∏—Ä–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏ —Å –∂–∞–Ω—Ä–∞–º–∏
features = df[['main_genre', 'side_genre']]

# –ü—Ä–∏–º–µ–Ω—è–µ–º One-Hot Encoding (pd.get_dummies) –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –±–∏–Ω–∞—Ä–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
# –≠—Ç–æ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_encoded = pd.get_dummies(features, prefix=['main_genre', 'side_genre'])

print("\n–†–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö –ø–æ—Å–ª–µ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è:", X_encoded.shape)
print("\n–§—Ä–∞–≥–º–µ–Ω—Ç –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–ø–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏):")
print(X_encoded.head(3))
```

```python
# --- 4. –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ (StandardScaler) ---

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
print("–î–∞–Ω–Ω—ã–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω—ã (StandardScaler).")

# --- 5. –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ (KNN) ---
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª–æ–≥–∏–∫—É NearestNeighbors –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø–æ—Ö–æ–∂–∏—Ö —Ñ–∏–ª—å–º–æ–≤
# n_neighbors=5: –±—É–¥–µ–º –∏—Å–∫–∞—Ç—å 5 –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π

model = NearestNeighbors(n_neighbors=5, metric='cosine')
model.fit(X_scaled)

print("\n–ú–æ–¥–µ–ª—å KNN –æ–±—É—á–µ–Ω–∞ –∏ –≥–æ—Ç–æ–≤–∞ –∫ —Ä–∞–±–æ—Ç–µ! üé¨")
```

```python
# --- 6. –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ ---
def recommend_movie(movie_index, n=5):
    """
    –ò—â–µ—Ç n —Ñ–∏–ª—å–º–æ–≤, –ø–æ—Ö–æ–∂–∏—Ö –Ω–∞ —Ñ–∏–ª—å–º —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∏–Ω–¥–µ–∫—Å–æ–º,
    –∏—Å–ø–æ–ª—å–∑—É—è –º–æ–¥–µ–ª—å NearestNeighbors.
    """
    # –ü–æ–ª—É—á–∞–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∏ –∏–Ω–¥–µ–∫—Å—ã n+1 –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (n+1, —Ç.–∫. 1-–π —ç—Ç–æ —Å–∞–º —Ñ–∏–ª—å–º)
    distances, indices = model.kneighbors([X_scaled[movie_index]], n_neighbors=n+1)

    try:
        # –í—ã–≤–æ–¥–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Ñ–∏–ª—å–º–∞
        original_title = df.iloc[movie_index]['Movie_Title']
        print(f"\n–ï—Å–ª–∏ –≤–∞–º –Ω—Ä–∞–≤–∏—Ç—Å—è —Ñ–∏–ª—å–º: {original_title}")
        print("–í–∞–º –º–æ–≥—É—Ç –ø–æ–Ω—Ä–∞–≤–∏—Ç—å—Å—è (—Ç–æ–ø 5):")

        # –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö —Å–æ—Å–µ–¥–µ–π, –Ω–∞—á–∏–Ω–∞—è —Å–æ –≤—Ç–æ—Ä–æ–≥–æ (–∏–Ω–¥–µ–∫—Å 1)
        for i in range(1, len(indices[0])):
            idx = indices[0][i]
            title = df.iloc[idx]['Movie_Title']
            main_g = df.iloc[idx]['main_genre']
            side_g = df.iloc[idx]['side_genre']
            # –î–∏—Å—Ç–∞–Ω—Ü–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç "–Ω–µ–ø–æ—Ö–æ–∂–µ—Å—Ç—å": —á–µ–º –±–ª–∏–∂–µ –∫ 0, —Ç–µ–º –±–æ–ª–µ–µ –ø–æ—Ö–æ–∂–∏ –∂–∞–Ω—Ä—ã
            distance = distances[0][i]
            print(f"{i}. {title} (–ñ–∞–Ω—Ä—ã: {main_g}, {side_g}) - –°—Ö–æ–∂–µ—Å—Ç—å (1 - –¥–∏—Å—Ç–∞–Ω—Ü–∏—è): {1 - distance:.3f}")

    except KeyError:
        print("\n‚ö†Ô∏è –û—à–∏–±–∫–∞: –ü—Ä–æ–≤–µ—Ä—å –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ —Å –∏–º–µ–Ω–∞–º–∏ —Ñ–∏–ª—å–º–æ–≤ –≤ —Ç–≤–æ–µ–º CSV.")
        print("–í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ 'Movie_Title'. –ï—Å–ª–∏ —É —Ç–µ–±—è –¥—Ä—É–≥–æ–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 'Title' –∏–ª–∏ 'Name'), –∑–∞–º–µ–Ω–∏ –µ–≥–æ –≤ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏.")
        print("–î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏:", list(df.columns))


movie_to_test = 50
print(f"–¢–µ—Å—Ç–∏—Ä—É–µ–º —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—é –¥–ª—è —Ñ–∏–ª—å–º–∞ —Å –∏–Ω–¥–µ–∫—Å–æ–º {movie_to_test}...")
recommend_movie(movie_to_test)
```

```python
# --- 7. –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é ---
def get_movie_recommendations_by_title(partial_title):
    """
    –ò—â–µ—Ç —Ñ–∏–ª—å–º –ø–æ —á–∞—Å—Ç–∏—á–Ω–æ–º—É –Ω–∞–∑–≤–∞–Ω–∏—é –∏ –≤—ã–¥–∞–µ—Ç —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏.
    """
    # 1. –ü–æ–∏—Å–∫ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ (—Ä–µ–≥–∏—Å—Ç—Ä–æ–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–π)
    # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ –∫–æ–ª–æ–Ω–∫–∞ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è 'Movie_Title'
    matches = df[df['Movie_Title'].str.contains(partial_title, case=False, na=False)]

    if len(matches) == 0:
        print(f"‚ùå –§–∏–ª—å–º —Å –Ω–∞–∑–≤–∞–Ω–∏–µ–º, —Å–æ–¥–µ—Ä–∂–∞—â–∏–º '{partial_title}', –Ω–µ –Ω–∞–π–¥–µ–Ω.")
        return

    # 2. –ï—Å–ª–∏ –Ω–∞—à–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ, –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º –≤—ã–±—Ä–∞—Ç—å –ø–µ—Ä–≤—ã–π –∏–ª–∏ —É—Ç–æ—á–Ω–∏—Ç—å
    if len(matches) > 1:
        print(f"üîé –ù–∞–π–¥–µ–Ω–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∏–ª—å–º–æ–≤ ({len(matches)}). –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–µ—Ä–≤—ã–π –∏–∑ —Å–ø–∏—Å–∫–∞:")
        print(f"   -> {matches.iloc[0]['Movie_Title']} (–ñ–∞–Ω—Ä—ã: {matches.iloc[0]['main_genre']}, {matches.iloc[0]['side_genre']})")

    else:
        print(f"‚úÖ –§–∏–ª—å–º –Ω–∞–π–¥–µ–Ω: {matches.iloc[0]['Movie_Title']}")

    # 3. –ü–æ–ª—É—á–∞–µ–º –∏–Ω–¥–µ–∫—Å –Ω–∞–π–¥–µ–Ω–Ω–æ–≥–æ —Ñ–∏–ª—å–º–∞
    movie_index = matches.index[0]

    # 4. –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä—É—é –º—ã –Ω–∞–ø–∏—Å–∞–ª–∏ –≤ –ë–ª–æ–∫–µ 5
    recommend_movie(movie_index)

# --- –ó–ê–ü–£–°–ö: –í–≤–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏—è ---
print("–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∏–ª—å–º–∞ (–Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º), –∫–æ—Ç–æ—Ä—ã–π –≤–∞–º –Ω—Ä–∞–≤–∏—Ç—Å—è:")
user_input = input("–ù–∞–∑–≤–∞–Ω–∏–µ —Ñ–∏–ª—å–º–∞: ") # –ü–æ–ª–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ—è–≤–∏—Ç—Å—è –ø–æ—Å–ª–µ –∑–∞–ø—É—Å–∫–∞

if user_input.strip():
    get_movie_recommendations_by_title(user_input)
else:
    print("–í—ã –Ω–∏—á–µ–≥–æ –Ω–µ –≤–≤–µ–ª–∏.")
```
